### 252 Days sliding window

The evolution of the first and second order statistics (mean and variance) is computed and displayed. They are calculated using a sliding window of 252 days with a 1 day increment. Due to the exponential behaviour of the stock market, the price volatility increases with the increase in value. However, in terms of percentage it does not. A stock loosing $100 in value was more relevant in 1930 than in 2017 but a loss of 10\% of the value is equaly important. The simple price index captures the exponential behaviour of the trend but fails to provide a realistic view of what the variance is when analyzing the whole century.

On the other hand, the log transform gives a linear representation of the growth of the SPX index and it makes the analysis much easier. The log transform of the SPX Index variance turns it into a stationary times-series where it can be seen when the economic downturns occured.

```python
# Sliding mean
plt.figure(dpi=100)
plt.subplot(211)
plt.title('252 day rolling mean of SPX Index data')
plt.xlabel('date')
plt.grid()
plt.plot(date_axis,spx.rolling(252).mean())

# Sliding standard deviation
plt.subplot(212)
plt.plot(date_axis,spx.rolling(252).var())
plt.title('252 day rolling variance of SPX Index data')
plt.xlabel('date')
plt.tight_layout()
plt.grid()
plt.show()

# Log Sliding mean
plt.figure(dpi=100)
plt.subplot(211)
plt.plot(date_axis,logpx.rolling(252).mean())
plt.title('252 day rolling mean of Log SPX Index data')
plt.xlabel('date')
plt.grid()

# Log Sliding standard deviation
plt.subplot(212)
plt.plot(date_axis,logpx.rolling(252).var())
plt.title('252 day rolling variance of Log SPX Index data')
plt.xlabel('date')
plt.tight_layout()
plt.grid()
plt.show()
```

### Suitability of log returns and the Jarque-Bera test

Log returns are of great importance in signal processing and are many times used over simple returns. Some of the reasons for this are its properties of time additivity (see section 1.1.5), mathematical tractability (exponents are easy to manipulate) and numerical stability (addition of small numbers is less prone to error than multiplication). Furthermore, log returns are less susceptible to outliers since any data-point that significantly deviates from the mean is compressed.

The Jarque-Bera test evaluates how well a gaussian model fits the time series being evaluated. It tests whether the kurtosis and skweness of the model is similar to that of a normal distribution. For its implementation, the scipy library is used. The p-value returned by the algorithm is 0. This means that the null hypothesis is rejected and the time-series does not follow a normal distribution.

```python
# Return the Jarque-Bera test p-value for a time-series x
p_value = stats.jarque_bera(spx)[1]
print(p_value)
```

### Simple and logarithmic returns

The listing below illustrates the function that computes the simple and log returns. One advantage of log returns is that they are time-additive. If an asset whose value is 1 increases to 2 and then goes back down to 1 then when using the simple return function, the purchaser of this asset does not have a return of 0 by the end of the day. However, if log returns are used, the final return is equal to zero.


```python
price = [1, 2, 1]

# Simple return
R = np.zeros(3)
R[0] = 0            # There is no return on the first day
for t in range(1,3):
    R[t] = (price[t]-price[t-1])/price[t-1]
print('Simple return:', R)
print('Purchaser ends up with:' + str(sum(R)) + ' $')

# Log return
r = np.zeros(3)
r[0] = 0            # There is no return on the first day
for t in range(1,3):
    r[t] = np.log(price[t]/price[t-1])
print('Logarithmic return:', r)
print('Purchaser ends up with:' + str(sum(r)) + ' $')
```

### Preference of simple returns over log returns

Although log returns are easier to manipulate mathematically. There are some reasons because of which simple returns can be preferred over log returns. When the time frame of analysis is large, the assumption of log-normality of the distribution does not hold because the skewness of financial data is negative and that of the model is positive. 

## ARMA vs. ARIMA Models for Financial Applications

An autoregresive moving average ARMA(p,q) process is a model used to predict stationary data. It consists of an AR part that autoregresses its own past values and an MA part that models the behaviour of the error based on previous error terms. The first intends to explain the momentum and reversion effects from trading markets and it's of order p-AR-lags. The later attempts to model the shock events present in financial data and it's of order q-MA-lags. This process requires the time-series to be stationary.

On the other hand, the autoregressive integrated moving average ARMA(p,d,q) is the same as ARMA but with the addition of a differencing component that makes the time-series stationary. 

### Log closing price of the S&P 500 from 2015 to 2019

The log transform of the S&P 500 time-series is displayed above. It can be observed that it is a nonstationary time-series. Because of this, the task of modeling the time-series is better suited for an ARIMA process than for an ARMA process.

```python
snp = pd.read_csv('snp_500_2015_2019.csv')
dates = snp['Date']
date_axis = [dt.datetime.strptime(d,'%Y-%m-%d').date() for d in dates]
snp_close = snp['Close'].to_frame().apply(np.log)
plt.figure(dpi=100)
plt.plot(date_axis, snp_close)
plt.xticks(rotation=45)
plt.title('Log closing price of S&P 500 from 2015 to 2019')
plt.xlabel('date')
plt.ylabel('log points')
plt.grid()
plt.show()
```

### Fit an ARMA(1,0)

An ARMA(1,0) is employed to predict on one day basis the S&P 500 during the period from 2015 to 2019. The resulting MSE can be seen below. 

```python
snp_arma = copy.deepcopy(snp_close)
snp_arma.columns = ['True']
model = ARIMA(snp_arma, order=(1,0,0))
result = model.fit()
snp_arma['Res'] = result.resid
snp_arma['Prediction'] = snp_arma['True'] - snp_arma['Res']
plt.figure(dpi=100)
plt.plot(date_axis, snp_close, 'blue')
print('Prediction MSE: ', np.mean((np.array(snp_arma['Prediction']) - np.array(snp_close)))**2)
plt.plot(date_axis, snp_arma['Prediction'], 'red',linewidth=0.8)
plt.xticks(rotation=45)
plt.title('ARMA prediction of log S&P 500 from 2015 to 2019')
plt.xlabel('date')
plt.ylabel('log points')
plt.legend(['True', 'ARMA(1,0) prediction'])
plt.grid()
plt.show()
```

The coefficients of the AR process are shown below. It can be obersved that there is a clear correlation between the present value and the inmediate past. 

```python
print(result.params)
print('The second value of the true time series is: ', snp_arma['True'][1])
print('The second value of the predicted time series is: ', snp_arma['Prediction'][1])
```

When calculating the coefficients of a higher order AR process, namely, AR(5) it becomes clear that a sample two time samples away from the present had barely no effect because the AR coefficients with index above one are nearly zero. Any AR model of higher order than 1 would be overfitting.

The use of an AR(1) for time-series forecasting is of very little practicality. The model takes the immediate past value and adds white noise to it. There is no useful information in the previous time values that can be used to forecast the future. Under a completely efficient market scenario, this model would be the best since any future values are just a random walk from the previous state.

```python
snp_arma = copy.deepcopy(snp_close)
snp_arma.columns = ['True']
params = ARIMA(snp_arma, order=(5,0,0)).fit().params

plt.figure(dpi=100)
plt.bar([1,2,3,4,5],np.array([params['ar.L' + str(i) +'.True'] for i in range(1,6)]))
plt.title('Coefficients of AR(5)')
plt.xlabel('coefficient')
plt.grid()
plt.show()
```

## Vector Autoregressive (VAR) Models

Vector autoregression (VAR) is a probabilistic process that models the linear interdependencies of different time-series. It's a generalization of the AR model for more than one variable.

### VAR Matrix form

The vector autoregressive model can be represented in matrix form as

\begin{equation}
\textbf{Y} = \textbf{B}\textbf{Z} + \textbf{U} \\
\end{equation}
where
\begin{equation}
    \textbf{Y} = 
    \begin{pmatrix}
        y_{1,T-1} & y_{1,T-2} & \cdots & y_{1,0} \\
        y_{2,T-1} & y_{2,T-2} & \cdots & y_{2,0} \\
        \vdots        &       \vdots  & \ddots &       \vdots \\
        y_{K,T-1} & y_{K,T-2} & \cdots & y_{K,0} \\ 
    \end{pmatrix}
\end{equation}


\begin{equation}
    \textbf{B} = 
    \begin{pmatrix}
        c_{1}      &      a_{1,1}^{1} & a_{1,2}^{1} & \cdots & a_{1,K}^{1} & a_{1,1}^{2} \cdots & a_{1,K}^{P} \\
        c_{2}      &      a_{2,1}^{1} & a_{2,2}^{1} & \cdots & a_{2,K}^{1} & a_{2,1}^{2} \cdots & a_{2,K}^{P} \\
        \vdots      &      \vdots        &       \vdots  & \ddots &       \vdots  &  \vdots \quad  \ddots & \vdots \\
        c_{K}      &      a_{K,1}^{1} & a_{K,2}^{1} & \cdots & a_{K,K}^{1} & a_{K,1}^{2} \cdots & a_{K,K}^{P} \\ 
    \end{pmatrix}
\end{equation}

\begin{equation}
    \textbf{Z} = 
    \begin{pmatrix}
            1             &        1      & \cdots &     1     \\
        y_{1,t-1}^{1} & y_{1,t-2}^{1} & \cdots & y_{1,t-p}^{1} \\
        y_{2,t-1}^{1} & y_{2,t-2}^{1} & \cdots & y_{2,t-p}^{1} \\
        \vdots        &       \vdots  & \ddots &       \vdots \\
        y_{K,t-1}^{1} & y_{K,t-2}^{1} & \cdots & y_{K,t-p}^{1} \\
        y_{1,t-1}^{2} & y_{1,t-2}^{2} & \cdots & y_{1,t-p}^{2} \\
        \vdots        &       \vdots  & \ddots &       \vdots \\
        y_{K,t-1}^{p} & y_{K,t-2}^{p} & \cdots & y_{K,t-p}^{p} \\ 
    \end{pmatrix}
\end{equation}

\begin{equation}
    \textbf{U} = 
    \begin{pmatrix}
        e_{1,T-1} & e_{1,T-2} & \cdots & e_{1,0} \\
        e_{2,T-1} & e_{2,T-2} & \cdots & e_{2,0} \\
        \vdots    &   \vdots  & \ddots &   \vdots \\
        e_{K,T-1} & e_{K,T-2} & \cdots & e_{K,0} \\ 
    \end{pmatrix}
\end{equation}



### Optimal set of coefficients for the matrix B

In order to minimize the error of the function above, an LSE estimator is used for the model. The cost function is given by:

\begin{equation}
J(\textbf{B}) = (\textbf{Y}-\textbf{B}\textbf{Z})^{T}(\textbf{Y}-\textbf{B}\textbf{Z})
\end{equation}

when expanded, it turns into

\begin{equation}
J(\textbf{B}) = \textbf{Y}\textbf{Y}^{T} - 2\textbf{Y}^{T}\textbf{B}\textbf{Z} + \textbf{Z}^{T}\textbf{B}^{T}\textbf{B}\textbf{Z}
\end{equation}

The aim is to minimize the cost function:

\begin{equation}
\frac{J(\textbf{B})}{d\textbf{B}} = 0
\end{equation}

This results in:

\begin{equation}
-2\textbf{Y}\textbf{Z}^{T} + \textbf{B}_{opt}(\textbf{Z}\textbf{Z}^{T}+\textbf{Z}\textbf{Z}^{T}) = 0 \\
\end{equation}
\begin{equation}
-2\textbf{Y}\textbf{Z}^{T} + 2\textbf{B}_{opt}\textbf{Z}\textbf{Z}^{T} = 0
\end{equation}

Finally, the optimal parameters are given by:

\begin{equation}
\textbf{B}_{opt} = \textbf{Y}\textbf{Z}^{T}(\textbf{Z}\textbf{Z}^{T})^{-1}
\end{equation}



### Stability of a VAR(1) process

A VAR(1) process is given by the following equation

\begin{equation}
\textbf{y}_{t} = \textbf{A}\textbf{y}_{t-1} + \textbf{e}_{t}
\end{equation}

For the system to be stable the eigenvalues of the matrix \textbf{A} must be smaller than one in absolute value. This is proven by taking the Z-transform:

\begin{equation}
Y(z) = \textbf{A}Y(z)z^{-1} + E(z)\\
\end{equation}
\begin{equation}
Y(z)(\textbf{I}-\textbf{A}z^{-1}) = E(z)\\
\end{equation}
\begin{equation}
H(z) = (\textbf{I}-\textbf{A}z^{-1})^{-1}\\
\end{equation}
\begin{equation}
\textbf{A} = \textbf{Q}\textbf{D}\textbf{Q}^{-1} \Rightarrow H(z) = (\textbf{I} - \textbf{Q}\textbf{D}\textbf{Q}^{-1}z^{-1})^{-1}\\
\end{equation}
\begin{equation}
\text{where } \textbf{A} \text{ has been diagonalized.}\\
\end{equation}

\begin{equation}
\text{For the system to be stable, there must be no poles. This happens when }\\
\end{equation}

|\lambda| < 1
\end{equation}

# Bond pricing

## Examples of bond pricing

An investor will be given 1100 in one year for an investment of 1000 now. It is of interest to calculate the percentage return with an annual, semi-annual, monthly and continuous period of compounding. These are given by the equations below:

\begin{equation}\label{eq:annual_compounding}
r_{1} = C_{i} \times (1+r_{f})\quad \text{Annual compounding} 
\end{equation}
\begin{equation}\label{eq:semiannual_compounding}
r_{1} = C_{i} \times (1+\frac{r_{f}}{2})^{2}\quad \text{Semi-annual compounding}  
\end{equation}
\begin{equation}\label{eq:monthly_compounding}
r_{1} = C_{i} \times (1+\frac{r_{f}}{12})^{12}\quad \text{Monthly compounding}  
\end{equation}
\begin{equation}\label{eq:continuous_compounding}
r_{1} = C_{i} \times e^{r_{f}}\quad \text{Continuous compounding}  
\end{equation}

Where \(r_{f}\) is the annual percentage return and \(C_{i}\) is the initial capital investment. Using these equations it can be calculated that, for the given scenario, the percentage return per annum is 10%, 9.76%, 9.56% and 9.53% for the annual, semi-annual, monthly and continuous compounding, respectively.

Using equations \ref{eq:monthly_compounding} and \ref{eq:continuous_compounding}, the relationship between monthly and continuous compounding is determined by:

\begin{equation}\label{eq:rel_month_cont_comp}
r_{f continuous} = ln((1+\frac{r_{f monthly}}{12})^{12})
\end{equation}

In this way, if the monthly compounding interest is of 15%, the equivalent continuous interest is 14.91%.

An account with a deposit of 10,000 dollars and an annual continuous compounding interest of 12% that is paid quarterly can be calculated using the equation below. It will result in quarterly payments of 318.74 dollars.

\begin{equation}\label{eq:quarterly_payment}
payment = \frac{1}{4}\times (e^{0.12}-10000)
\end{equation}




## Forward rates

If a one year interest is of 5% and a two year interest is of 7% for an investment of 100, the forward rate of interest equals 9%. This is the extra capital obtained on the initial investment over the two year period using the second investment strategy. The forward price can be calculated using the following equation:

\begin{equation}\label{eq:forward_price}
F = S_{0}e^{(r-q)T}
\end{equation}

Where \(F, S_{0}, r, q\) are the forward price, initial investment, interest rate and the cost of carry, respectively. With the current value of the dollar, and assuming no cost of carry, it makes little sense to make a 2 year 7% interest investment because in absolute values, the difference is very small and investment time period is of two years instead of one. If the initial investment capital was of a higher order, it would make more economic sense. They only reason to invest such a small amount of money is to avoid the devaluation caused by inflation or to reduce the risk of storing the capital somewhere else. It would, therefore, be more practical to make use of the first investment strategy with an annual interest rate of 5%. This gives the investor the flexibility to get back the money in a shorter amount of time and to decide whether or not to reinvest. If then cost of carry is assumed to be of 4%, both investments would give the same return but the second strategy would take twice the amount of time. Moreover, an assumption was made that the investment was free of risk. This is often not the case and it must be taken into account when selecting the investment strategy.

For the second strategy to become the best option several certain requirements are need to be met. First, the annual inflation figures and risk of default must be known in order to aid in the selection of a cost of carry. Then the interest rate must be larger than this value and considerably larger than that of the first strategy for the reasons mentioned above.


## Duration of a coupon-bearing bond

The duration of a bond is the weighted average of the times of each to the payments until the maturity date and it is described in the equation below. The duration is used to measure the sensitivity of any given bond to a change in the interest rates. Using this equation, the duration of a 1% bond can be found to be equal to 6.76.

\begin{equation}\label{eq:duration}
Duration = \sum_{i=1}^{T}\frac{t_{i}\times PV(C_{i})}{PV}
\end{equation}

A modified version of the duration is the volatility and it is used to measure the percentage change of the bond price with a one percent change in yield. It is defined in equation \ref{eq:modified_duration}. Using this expression, the modified duration is equal to 6.32. 

\begin{equation}\label{eq:modified_duration}
\text{Modified duration}  = \text{volatiltiy (%)} = \frac{Duration}{1+yield}
\end{equation}

While the first expression for the duration is the weighted average time until repayment, the modified duration measures the sensitivity of the bond price with respect to the yield.

Bond prices are susceptible to interest rates and have an inverse relationship. Therefore, when selecting a bond to carry the pension plan of an investor it is useful to understand how susceptible it is to changes in the interest rates. The modified duration can be used for this purpose. This information can also help determine whether to agree to a fixed or a floating interest rate. The first would have a higher relative yield in the case of the interest rates being lowered, and conversely, and the latter would fluctuate with them. This is, therefore, a risk/yield trade-off that the investor can for his/hers pension plan.

### Rolling beta

It is of interest to study the effect of the expected excess asset returns to the expected excess market returns. If the market fluctuates, it can be estimated how individual assets will behave which will aid in the evaluation of risk. This is done through the calculation of \(\beta_{i}\), which is the ratio of the covariance of the individual asset and the market return over the variance of the return of the market. This parameter is calculated for the market currently under study. As it can be observed in the figure below, this parameter has a high fluctuation and cannot be assumed constant. The higher the \(\beta_{i}\), the more the returns will fluctuate when the market is volatile. On the other hand, if the beta is positive an smaller than 1, the value of the asset will tend to remain stable when the market fluctuates. A negative beta occurs when an asset tends to have an opposite behavior to the market. The selection of a portfolio is greatly determined by this parameter. 

```python
plt.figure(dpi=100)
plt.title('Estimated rolling Beta for the 157 European companies')
plt.xlabel('date')
plt.ylabel('estimated Beta')
plt.grid()
betas = df['ret'].rolling(22).cov(avg_returns).dropna()/np.var(avg_returns)
plt.plot(betas.index, betas)
plt.xticks([betas.index[i] for i in np.linspace(0,len(betas.index)-1, 10).astype(int)],rotation=45)
plt.tight_layout()
plt.show()
```

# Portfolio Optimization

## Adaptive minimum-variance portfolio optimization

### Derivation of the minimum-variance portfolio

An investor will look to minimize the variance of his/her portfolio in order to reduce the risk of the overall investment. This can be achieved by selecting a set of assets whose covariance is low. In this exercise, the minimum variance portfolio is studied. This is an optimization problem whose objective function minimizes the portfolio variance subject to the condition of the sum of the weights to be equal to one. The equation is shown below:

\begin{equation}\label{eq:}
min_{\textbf{w}} \frac{1}{2}\textbf{w}^{T}\textbf{C}\textbf{w}
\end{equation}
subject to
\begin{equation}\label{eq:}
\textbf{w}^{T}\textbf{1} = 1
\end{equation}

Using the lagrange multipliers the problem to solve becomes:

\begin{equation}\label{eq:}
min_{\textbf{w},\lambda} J'(\textbf{w},\lambda, C) = \frac{1}{2}\textbf{w}^{T}\textbf{C}\textbf{w} - \lambda(\textbf{w}^{T}\textbf{1} - 1)
\end{equation}

By taking the gradient with respect to \(\textbf{w}\) and \(\lambda\) a system of two equations is constructed.

\begin{equation}\label{eq:system_eq_1}
\textbf{w}\textbf{C} - \lambda \textbf{1}^{T}
\end{equation}
\begin{equation}\label{eq:system_eq_2}
\textbf{w}^{T}\textbf{1} - 1 = 0
\end{equation}

Rearranging Eq. \ref{eq:system_eq_1} leads to

\begin{equation}\label{eq:}
\textbf{w}^{T} = \lambda\textbf{1}^{T}\textbf{C}^{-1}\\
\end{equation}
\begin{equation}\label{eq:}
\textbf{w}^{T}\textbf{1} = \lambda\textbf{1}^{T}\textbf{C}^{-1}\textbf{1}
\end{equation}


And using Eq. \ref{eq:system_eq_2} converts the equation into:

\begin{equation}
1 = \lambda\textbf{1}^{T}\textbf{C}^{-1}\textbf{1}\\
\end{equation}

Solving for \(\lambda\) results in:

\begin{equation}\label{eq:solution_lambda}
\lambda = \frac{1}{\textbf{1}^{T}\textbf{C}^{-1}\textbf{1}}
\end{equation}

Finally, substituting Eq.\ref{eq:solution_lambda} into Eq.\ref{eq:system_eq_1} gives the result for the optimal weights:

\begin{equation}\label{eq:solution_w}
\textbf{T}  = \frac{\textbf{1}^{T}\textbf{C}^{-1}}{\textbf{1}^{T}\textbf{C}^{-1}\textbf{1}}
\end{equation}

When employing the optimal weights in Eq.\ref{eq:solution_w} for the construction of a portfolio, the mean and variance of its set of securities become:

\begin{equation}\label{eq:}
\hat{\mu} = (\frac{\textbf{1}^{T}\textbf{C}^{-1}}{\textbf{1}^{T}\textbf{C}^{-1}\textbf{1}})^{T}\textbf{\mu} 
\end{equation}
\begin{equation}\label{eq:}
\hat{\sigma}^{2} = \frac{\textbf{1}^{T}}{\textbf{1}^{T}\textbf{C}^{-1}\textbf{1}}(\frac{\textbf{1}^{T}\textbf{C}^{-1}}{\textbf{1}^{T}\textbf{C}^{-1}\textbf{1}})^{T}
\end{equation}




### Performance of a portfolio with two strategies

In this exercise, two different strategies are employed: the minimum variance portfolio and the equally weighted portfolio. The first constructs a portfolio using Eq.\ref{eq:solution_w}. This way, the volatility of the portfolio is minimized. However, this does not mean that returns will be higher. The second strategy gives the same weighting to all the assets in the market, with the sum being equal to one. In order to select the weights, a training dataset is constructed with the first half of the whole time series. The rest of the dataset is used for testing the portfolio selection. Other strategies for the train/test split could have been used such as a random selection of the datapoints. However, it as assumed that the investor will be looking at the whole time series from the past whenever creating his/her portfolio.

The results are shown below. It can be observed that variance is significantly reduced when using a minimum variance portfolio over an equally weighted one. Although both cumulative returns are negative, the ones from the minimum variance portfolio are slightly better. The figure below displays the time series of the cumulative return of both strategies. For half the time period, the equally weighted strategy yielded positive returns until it quickly decreased. Om the other hand, the minimum variance strategy consistently yielded negative returns but with a lower volatility. The figure illustrates the aim of the minimum variance strategy i.e reduce the volatility. Since the optimization problem had no incentive to look for assets whose returns where positive, this strategy does not necessarily lead to higher returns.

```python
df = pd.read_csv(r'fsp_case_31_BSD.csv',index_col=0,header=[0,1]).fillna(0)
ret = np.array(df['ret']).T
train_r = ret[:,:int(521/2)]
test_r = ret[:,int(521/2):]

C = np.cov(train_r)
C_inv = np.linalg.inv(C)
ones = np.ones(C.shape[0]).reshape(-1,1)

w_opt = ((np.dot(ones.T,C_inv))/np.dot(np.dot(ones.T,C_inv),ones)).T
w_2 = np.ones(C.shape[0])*(1/C.shape[0])

r_opt = np.array([np.dot(w_opt.T,test_r[:,n]) for n in range(test_r.shape[1])])
r_2 = np.array([np.dot(w_2.T,test_r[:,n]) for n in range(test_r.shape[1])])
R_opt = np.cumsum(r_opt)
R_2 = np.cumsum(r_2)
print('The return of the minimum variance weights is ', R_opt[-1])
print('The variance of the minimum variance weights is ', np.var(r_opt))
print('The return of the equally weighted portfolio is ', R_2[-1])
print('The variance of the equally weighted portfolio is ', np.var(r_2))

plt.figure(dpi=100)
plt.plot(dates[int(521/2):],R_opt)
plt.plot(dates[int(521/2):],R_2)
plt.xticks([dates[int(i)] for i in list(np.linspace(int(521/2),520,15))], rotation=45)
plt.grid()
plt.xlabel('date')
plt.ylabel('cumulative return')
plt.title('Cumulative return of the optimal and equal weights portfolios')
plt.legend(['MV weights', 'Equal weights'])
plt.tight_layout()
plt.show()
```

This way, the the optimal weights can be recomputed at every time step depending on how the securities are fluctuating. Intuitively, this approach should yield a higher performance than the static MV and equally weighted portfolios because the algorithm could correct for any changes in the market. However, it would also only be able to use a smaller amount of data-points which is determined by the window size.

## Robust Estimators

In this section three estimators (i.e median, interquartile range and median absolute deviation) will be evaluated. They are characterized by being robust. This means that it is more difficult for the estimators to be wrong in the presence of outliers if the the sample distribution is gaussian. 

### Estimator functions

Three functions are written and each of them computes the above mentioned estimators. These can be seen in the listing below.


```python
def median(df):
    return df.median()

def iqr(df):
    return np.array(df.quantile([0.25,0.75]))

def mad(df):
    return robust.mad(df.values)
```

### Efficiency

In this exercise, the statistical efficiency of each of the estimators will be assessed by investigating how many samples an estimator requires to obtain the theoretical result up until the third decimal. Each of the estimators evaluates an array of size \(N\) containing random samples from a gaussian distribution of mean 0 and unity standard deviation. For each value of \(N\), the mean of 10,000 iterations is computed. If this value is not accurate to the third decimal, \(N\) is increased by 10. The final \(N\) will tell the number of samples that are required on average to compute an accurate result. Using the code below it can be observed that on average, the median, IQR and MAD estimators require 1260, 2280 and 1800 samples in order to be sufficiently accurate to the third decimal. From these findings it can be said that the most efficient estimator is the median, followed by the MAD and IQR. This result disagrees with [1](http://web.ipac.caltech.edu/staff/fmasci/home/astro_refs/BetterThanMAD.pdf), which says that the least efficient estimator is MAD. 


```python
df = pd.DataFrame()
M = 10000
N = 200
np.random.seed(1234)
df['normal'] = np.random.normal(0,1,N)

estimate = 9999
while round(estimate,3) != 0:
    df = pd.DataFrame()
    df['normal'] = np.random.normal(0,1,N)
    for i in range(M):
        estimate = estimate + median(df['normal'][:N])
    estimate = estimate / M
    N = N + 20
N_median = N - 20
print('It takes ' + str(N_median) + ' samples for the median estimator to be accurate')

N = 200
estimate = [9999,9999]
while (round(estimate[0],3) != -0.67) and (round(estimate[1],3) != 0.67):
    df = pd.DataFrame()
    df['normal'] = np.random.normal(0,1,N)
    for i in range(M):
        estimate = estimate + iqr(df['normal'][:N])
    estimate = estimate / M
    N = N + 20
N_iqr = N - 20
print('It takes ' + str(N_iqr) + ' samples for the IQR estimator to be accurate')

N = 200
estimate = 9999
while round(estimate,3) != 1:
    df = pd.DataFrame()
    df['normal'] = np.random.normal(0,1,N)
    for i in range(M):
        estimate = estimate + mad(df['normal'][:N])
    estimate = estimate / M
    N = N + 20
N_mad = N - 20
print('It takes ' + str(N_mad) + ' samples for the MAD estimator to be accurate')
```

Additionally, the computational efficiency of the three estimators will be evaluated using the IBM closing stock price from January 2010 to April 2019. The figure below displays the amount of time it takes for each of the functions to compute the estimator for N number of samples. Each observation is averaged over ten iterations in order to provide a more reliable result.

The fastest estimator function is the median with an approximate computation time of 7 \(\mu s\). This function is followed then by the MAD function, which takes approximately 12 \(\mu s\). Finally, the slowest estimator function is the IQR. This function takes an average of 25 \(\mu s\) to compute. It can be observed that there is a slight linear relationship between the number of samples N and the computation time. It is to be expected that if the number of samples grows, so will the time it takes for the estimator to be computed. 

```python

ibm3 = web.DataReader("ibm", 'yahoo')

median_time = np.zeros(len(ibm3['Close']))
iqr_time = np.zeros(len(ibm3['Close']))
mad_time = np.zeros(len(ibm3['Close']))

for i in range(0,10):
    for n in range(1,len(ibm3['Close'])+1):
        time1 = time.time()
        median(ibm3['Close'][:n])
        time2 = time.time()
        iqr(ibm3['Close'][:n])
        time3 = time.time()
        mad(ibm3['Close'][:n])
        time4 = time.time()
        median_time[n-1] = median_time[n-1] + time2 - time1
        iqr_time[n-1] = iqr_time[n-1] + time3 - time2
        mad_time[n-1] = mad_time[n-1] + time4 - time3

median_time = 1000*median_time/len(ibm3['Close'])
iqr_time = 1000*iqr_time/len(ibm3['Close'])
mad_time = 1000*mad_time/len(ibm3['Close'])

plt.figure()
plt.plot(range(1,len(median_time)+1), median_time)
plt.plot(range(1,len(iqr_time)+1), iqr_time)
plt.plot(range(1,len(mad_time)+1), mad_time)
plt.title('Computation time for three estimators')
plt.xlabel('N')
plt.ylabel('time (ms)')
plt.legend(['Median', 'IQR', 'MAD'])
plt.grid()
plt.show()
```

### Breakdown point of the estimators

The breakdown point of an estimator is the percentage of outliers that can be present in the sample data without making the estimator give an arbitrarily wrong value. It is a test to the robustness of any estimator. Since it is required for a corrupt sample size to be less than 50% of the total data in order to be classified as an outlier, it is also the maximum breakdown point. Two types of breakdown points can be studied, the finite sample and the asymptotic breakdown point [1](http://www.stat.umn.edu/geyer/5601/notes/break.pdf). The first is the breakdown point of the function with \(N\) number of samples and the latter has the \(N\) tend to infinity. 

For the median estimator, if a minority of data-points are outliers, the finite sample breakdown point is given by:

\begin{equation}\label{eq:finite_b_median}
floor(\frac{n-1}{2n})
\end{equation}

where \(n\) is the number of samples. For the asymptotic breakdown point, this function is equal to 0.5. As robust as the estimator can be.

The IQR estimator is a translated version of two medians for the first and third quartiles. For this reason, the number of outliers needed to give an arbitrarily wrong estimate is half than that of the median, i.e 0.25 breakdown point.

In order to test these theoretical results, an array of size \(N\), whose elements belong to a gaussian distribution with zero mean and unity standard deviation, are stacked to another array of size \(N-n\) with arbitrarily large values. The total size of the array is then \(2N-n\). With \(n\) varying from 0 to \(N\), each of the estimator functions computes the input array. If the estimate is within a reasonable distance to the theoretical result, the estimator has neglected the outliers of the distribution. This way it can be determined how many samples are needed to corrupt the estimate of the distribution and the breakdown point can be evaluated. The results below show that the breakdown points of the median, IQR and MAD estimators are 0.5, 0.25 and 0.49, respectively. These results agree with the theoretical estimates derived in this section. 


```python
N = 2000
deviation = 5
th_estimate = 0
gauss = np.random.normal(0,1,N)
for n in range(0,N):
    dist = pd.DataFrame()
    dist['stack'] = np.hstack((gauss, np.ones(N-n)*1000))
    estimate = median(dist['stack'])
    if estimate < (th_estimate + deviation):
        print('Breakdown point of the median estimator is', str(round((N-n)/(2*N-n),2)), '.')
        break
        
th_estimate = [-0.67, 0.67]
gauss = np.random.normal(0,1,N)
for n in range(0,N):
    dist = pd.DataFrame()
    dist['stack'] = np.hstack((gauss, np.ones(N-n)*1000))
    estimate = iqr(dist['stack'])
    if estimate[0] > (th_estimate[0] - deviation) and estimate[1] < (th_estimate[1] + deviation):
        print('Breakdown point of the IQR estimator is ', str(round((N-n)/(2*N-n),2)), '.')
        break
        
th_estimate = 1
gauss = np.random.normal(0,1,N)
for n in range(0,N):
    dist = pd.DataFrame()
    dist['stack'] = np.hstack((gauss, np.ones(N-n)*1000))
    estimate = mad(dist['stack'])
    if estimate < (th_estimate + deviation):
        print('Breakdown point of the MAD estimator is ', str(round((N-n)/(2*N-n),2)), '.')
        break
```

### Performance of the regression with outliers

The results from the previous two exercises show that the R squared error is better (higher) for all of the regressed stocks when using OLS rather than Huber. This is due to the fact that there are not many real outliers in the dataset being studied. In order to study the error when more outliers are present, an outlier ratio is introduced by multiplying that proportion of the data by random integers in the range of 1 to 5. 

The R squared performance of the regressors is evaluated for different values of the outlier ratio. The results show that the Huber regression does not perform as well as the OLS regression for any outlier ratio. One explanation for this could be that the outliers present a symmetric distribution around the true value and the loss introduced by any of the outliers is compensated by its mirrored version. Moreover, the larger the outlier ratio, it is clear to see that the performance of the regression is worse.  

```python

np.random.seed(1234)
y = dji['Ret']
score_ols = np.zeros((3,50))
score_hub = np.zeros((3,50))
n = 0
for outlier_ratio in range(0,50):
    outlier_ratio = outlier_ratio / 100
    i = 0
    np.random.seed(1234)
    if outlier_ratio > 0:
        random_idxs = np.random.random_integers(1,len(ibm['Ret'])*outlier_ratio,int(len(ibm['Ret'])*outlier_ratio))
    outlier_mult = np.random.random_integers(1,5,int(len(ibm['Ret'])*outlier_ratio))
    ols = LinearRegression()
    hub = HuberRegressor()
    companies = ['AAPL', 'IBM', 'JPM']
    for X in [np.array(aapl['Ret']).reshape(-1,1), np.array(ibm['Ret']).reshape(-1,1), np.array(jpm['Ret']).reshape(-1,1)]:
        if outlier_ratio > 0:
            for k in range(len(random_idxs)):
                X[random_idxs[k]] = X[random_idxs[k]]*outlier_mult[k]
        ols = ols.fit(X, y)
        hub = hub.fit(X,y)
        score_ols[i,n] = round(ols.score(X,y),5)
        score_hub[i,n] = round(hub.score(X,y),5)
        i = i + 1
    n = n + 1

plt.figure(dpi=100)
handlers = [None]*6
for n in range(0,3):
    handlers[2*n], = plt.plot(np.array(range(50))/100,score_ols[n],label=str(companies[n]+' OLS'))
    handlers[2*n+1], = plt.plot(np.array(range(50))/100,score_hub[n],label=str(companies[n] + ' Huber'))
plt.grid()
plt.title('R squared error for OLS and Huber regression')
plt.xlabel('Outlier ratio')
plt.ylabel('R squared error')
plt.legend(handles=handlers)
plt.show()
```
